### Ouliers

```python
df_outliers = df_clean.copy()
```

```python
num = df_clean.select_dtypes(exclude='object').columns
```

```python
# Melakukan identifikasi kolom yang mengalami outliers
print(f'Jumlah baris: {len(df_outliers)}')

outlier = []
no_outlier = []
is_outlier = []
low_lim = []
high_lim = []


filtered_entries = np.array([True] * len(df_outliers))
for col in num:
    Q1 = df_outliers[col].quantile(0.25)
    Q3 = df_outliers[col].quantile(0.75)
    IQR = Q3 - Q1
    low_limit = Q1 - (IQR * 1.5)
    high_limit = Q3 + (IQR * 1.5)

    #mulai filter outlier
    filter_outlier = ((df_outliers[col] >= low_limit) & (df_outliers[col] <= high_limit))
    outlier.append(len(df_outliers[~filter_outlier]))
    no_outlier.append(len(df_outliers[filter_outlier]))
    is_outlier.append(df_outliers[col][~filter_outlier].any())
    low_lim.append(low_limit)
    high_lim.append(high_limit)

    filtered_entries = ((df_outliers[col] >= low_limit) & (df_outliers[col] <= high_limit)) & filtered_entries

print("Outlier All Data :", len(df_outliers[~filtered_entries]))
print("Not Outlier All Data :", len(df_outliers[filtered_entries]))
print()

pd.DataFrame({
    "Column Name":num,
    "is Outlier": is_outlier,
    "Lower Limit": low_lim,
    "Upper Limit": high_lim,
    "Outlier":outlier,
    "No Outlier":no_outlier
})
```

**Observasi:**  
Terdapat banyak data numerik yang mengalami outlier sehingga perlu dilakukan handle Outliers, saya menggunakan metode IQR utnuk mengatasi hal tersebut.

```python
# Melakukan handle outlier dengan menggunakan metode IQR
columns_is_outlier = df_outliers[num].drop(['loan_amnt', 'loan_target'], axis=1)

print('Jumlah baris sebelum handle outlier :', df_outliers.shape[0])
# Iterasi melalui setiap kolom untuk menangani outlier
for column in columns_is_outlier:
    Q1 = df_outliers[column].quantile(0.25)
    Q3 = df_outliers[column].quantile(0.75)
    IQR = Q3 - Q1

    # Menghitung batas bawah dan atas outlier
    lower_limit = Q1 - (1.5 * IQR)
    upper_limit = Q3 + (1.5 * IQR)

    # Menghapus outlier dari DataFrame
    df_outliers = df_outliers[(df_outliers[column] >= lower_limit) & (df_outliers[column] <= upper_limit)]

print('Jumlah baris setelah handle outlier :', df_outliers.shape[0])
```

```python
def evaluate_model(model, prediction, X_train, y_train, X_test, y_test):
    # Calculate metrics for Data Train
    train_accuracy = accuracy_score(y_train, prediction[0])
    train_precision = precision_score(y_train, prediction[0])
    train_recall = recall_score(y_train, prediction[0])
    train_f1 = f1_score(y_train, prediction[0])
    train_roc_auc = roc_auc_score(y_train, prediction[0])

    # Print the metrics for Data Train
    print("Metrics for Data Train:")
    print(f"Accuracy: {train_accuracy:.2f}")
    print(f"Precision: {train_precision:.2f}")
    print(f"Recall: {train_recall:.2f}")
    print(f"F1 Score: {train_f1:.2f}")
    print(f"ROC AUC: {train_roc_auc:.2f}")

    # Calculate metrics for Data Test
    test_accuracy = accuracy_score(y_test, prediction[1])
    test_precision = precision_score(y_test, prediction[1])
    test_recall = recall_score(y_test, prediction[1])
    test_f1 = f1_score(y_test, prediction[1])
    test_roc_auc = roc_auc_score(y_test, prediction[1])

    # Print the metrics for Data Test
    print("\nMetrics for Data Test:")
    print(f"Accuracy: {test_accuracy:.2f}")
    print(f"Precision: {test_precision:.2f}")
    print(f"Recall: {test_recall:.2f}")
    print(f"F1 Score: {test_f1:.2f}")
    print(f"ROC AUC: {test_roc_auc:.2f}")
```
